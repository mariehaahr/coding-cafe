```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
data = pd.read_csv("Noisy_Bootstrap_Vin_datasaet.csv", delimiter=";")
data
```

```python
def convert(column_name):
    l = []
    l2 = []
    for i in data[column_name]:
        if len(i) > 6:
            new = float(i[0:5])
            l.append(float(new))
        else:
            l.append(float(i))
    for j in l:
        noise = random.randint(-3,3) + round(random.random(),2)
        if j > np.average(l) + 10:
            j2 = np.average(l) + noise
            l2.append(j2)
        else:
            l2.append(j)

    return l2
```

```python
alc = convert("alcohol")
mal = convert("malic_acid")
data["new_alc"] = alc
data["new_mal"] = mal

plt.scatter(data["new_alc"], data["new_mal"], c=data["wine_label"])
```

```python
data = pd.read_csv("new_data.csv")
data
```

```python
def add_noise(column_name):
    l = []
    for i in data[column_name]:
        noise = random.randint(-3,3) + round(random.random(),3)
        new = i + noise
        l.append(new)
    return l
    
```

```python
data["new_alc"] = add_noise("alcohol")
data["new_mal"] = add_noise("malic_acid")
```

```python
fig, ax = plt.subplots(ncols=2, nrows=1, figsize=(15,7))
ax[0].scatter(data["new_alc"], data["new_mal"], c=data["wine_label"], s=5)
ax[1].scatter(data["alcohol"], data["malic_acid"], c=data["wine_label"], s=5)
```
